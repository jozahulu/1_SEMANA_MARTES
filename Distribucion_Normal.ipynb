```python
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd

# definicion de parametros de la distribucion normal
mu = 0
sigma = 5
rng = range(-30,30)

# Generalmente la distribucion normal esta dada por valores 
dist = norm(mu, sigma)

#Trazar la función de densidad de probabilidad y de esta distribución.
# el método pdf () toma toma en una lista x valores y devuelve una lista de y.

plt.subplot(5,1,1) # Crea una cuadrícula de parcelas de 3 filas y 1 columna y representa el siguiente gráfico en el espacio1.
plt.plot(rng, dist.pdf(rng), 'r', linewidth=2)
plt.title('probabilidad de la distribucion normal')


# Trazar la función de densidad de probabilidad y de esta distribución.
plt.subplot(5,1,3)
plt.plot(rng, dist.cdf(rng))
plt.title('Función de distribución cumulutativa de distribución normal')

# Dibuja 1000 muestras de la variable aleatoria.
sample = dist.rvs(size=10000)

print ("Muestra de estadísticas descriptivas:")
print( pd.DataFrame(sample).describe())

# Trace un histograma de la muestras.
plt.subplot(5,1,5)
plt.hist(sample, bins=100, normed=True)
plt.plot(rng, dist.pdf(rng), 'r--', linewidth=2)
plt.title('10,000 muestras aleatorias de distribución normal')

# Muestra las graficas.
plt.show()
```
