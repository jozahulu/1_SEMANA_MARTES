```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
rad = 15   # Angulos en radianes
N = 10000   # Numero de datos aleatorios

angulo = np.random.uniform(0.0, 2.0*np.pi, N)
r = rad * np.random.uniform(0.0, 1.0, N) # radio del circulo
# Coordenadas polares 
x = r * np.cos(angulo)
y = r * np.sin(angulo)
plt.title('El Circulo ')
plt.plot(x, y, "green", ms=1)
plt.axis([-25, 25, -25, 25])
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import mpl_toolkits.mplot3d
import matplotlib.pyplot as pp
```

```python
N= 6000 # Numero de Datos
rad = 9000000     # Numero de Datos para los angulos
phi = np.random.uniform(0.0, 2.0*np.pi, N)     # Represnta el angulo Pi
angulo = np.random.uniform(0.0, 1.0*np.pi, N)
r = rad * np.random.uniform(0.0, 1.0, N)
 # Representan las coordenadas esfericas

x = r * np.sin(angulo) * np.cos(phi)  
y = r * np.sin(angulo) * np.sin(phi)
z = r * np.cos(angulo)

```

```python
pp.figure().add_subplot(111, projection='3d').scatter(x, y, z,color='green');
pp.show()
```
